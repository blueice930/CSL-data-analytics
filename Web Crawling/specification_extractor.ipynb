```python
import pandas as pd
import re
```

```python
pd.read_csv('./data/HKTVMall/Products_HKTVMall.csv', index_col=0)
```

```python
df = pd.read_csv('./data/HKTVMall/Products_HKTVMall.csv', index_col=0)
index_list = df["Product Index"].values.tolist()
name_list = df["Product Name"].values.tolist()
price_list = df["Price"].values.tolist()
```

```python
spec_list = []
for name in name_list:
    spec_str = "+".join(re.findall("[0-9]+[GM]B?", name))
    if len(spec_str) == 0:
        spec_str = "None"
    spec_list.append(spec_str)
```

```python
store_dict = {
    "Product Index": index_list,
    "Product Name": name_list,
    "Product Specification": spec_list,
    "Product Price": price_list
}
p = pd.DataFrame(store_dict)
```

```python
p.to_csv("./data/specification/HKTVMall_specification", sep=',', na_rep='N/A', encoding="utf_8_sig")
```
