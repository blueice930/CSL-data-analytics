```python
import pandas as pd
import re
```

```python
pd.read_csv('../Data Analysis/Updated-CSV/all_info_data_suning.csv', index_col=0)
```

```python
df = pd.read_csv('../Data Analysis/Updated-CSV/all_info_data_suning.csv', index_col=0)
index_list = df["product_index"].values.tolist()
name_list = df["product_name"].values.tolist()
brand_num_list = df["brand_to_num"].values.tolist()
price_list = df["price"].values.tolist()
mean_rate_list = df["mean_rates"].values.tolist()
total_comment_list = df["total_comment"].values.tolist()
```

```python
spec_list = []
for name in name_list:
    spec_str = "+".join(re.findall("[0-9]+[GM][B\s]", name))
    if len(spec_str) == 0:
        spec_str = "None"
    spec_list.append(spec_str)
spec_list
```

```python
ram_list = []
storage_list = []
for spec in spec_list:
    if spec == "None":
        ram_list.append(0)
        storage_list.append(0)
    else:
        s = re.findall("[0-9]+", spec)
        if len(s) > 2: # needs manual confirmation
            ram_list.append(-1)
            storage_list.append(-1)
        elif len(s) == 1: 
            ram_list.append(0)
            storage_list.append(int(s[0]))
        else:
            s = [int(i) for i in s]
            ram_list.append(min(s))
            storage_list.append(max(s))
```

```python
store_dict = {
    "index": index_list,
    "name": name_list,
    "ram": ram_list,
    "storage": storage_list,
    "brand_in_num": brand_num_list,
    "mean_rate": mean_rate_list,
    "total_comment": total_comment_list,
    "price": price_list
}
p = pd.DataFrame(store_dict)
p
```

```python
p.to_csv("./data/specification/Suning_specification.csv", sep=',', na_rep='N/A', encoding="utf_8_sig")
```
